```python
raw_address=usaddress.parse(street_example)
raw_address
```

```python
dict_address={}
for i in raw_address:
    dict_address.update({i[1]:i[0]})
dict_address
```

```python
dict_address['StreetName']+','+dict_address['AddressNumber']
```

```python
def cleaning_1(raw):
    raw_address=usaddress.parse(raw)
    dict_address={}
    for i in raw_address:
        dict_address.update({i[1]:i[0]})
    clean_adress=dict_address['AddressNumber']+","+str(dict_address['StreetName'])
    return clean_adress
```

```python
rest.address.sample(5).apply(cleaning_1)
```

```python
def cleaning_check(raw):
    raw_address=usaddress.parse(raw)
    dict_address={}
    for i in raw_address:
        dict_address.update({i[1]:i[0]})
        
    if 'StreetName' in dict_address.keys() and 'AddressNumber' in dict_address.keys():
        clean_adress=dict_address['AddressNumber']+","+str(dict_address['StreetName'])
        return clean_adress
    else:
        return 'no street or number'
```

```python
rest['clean_street_check']=rest.address.apply(cleaning_check)
```

```python
def cleaning_final(raw):
    if raw.startswith('OLVERA'):
        clean_adress='OLVERA,Los Angeles,USA'
    elif raw.startswith('1033 1/2 LOS ANGELES'):
        clean_adress='1033 1/2 LOS ANGELES ST,Los Angeles,USA'
    else:
        raw_address=usaddress.parse(raw)
        dict_address={}
        for i in raw_address:
            dict_address.update({i[1]:i[0]})
        clean_adress=dict_address['AddressNumber']+" "+str(dict_address['StreetName'])+str(', Los Angeles,USA')
    return clean_adress
```

```python
address_list = rest["address"].to_list()
street_list = []
for addr in address_list:
    try:
        data = usaddress.tag(addr)
        if "StreetName" in data[0].keys():
            street_list.append(data[0]["StreetName"])
        else:
            street_list.append("no city/town")
    except:
        if "ASTRONAUT" in addr:
            street_list.append("ASTRONAUT")
        else:
            street_list.append("no street address")

rest["street"] = street_list
rest["street"] = rest["street"].str.title()
rest.head()
```
